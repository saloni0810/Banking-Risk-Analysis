```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns

for col in categorical_cols:
    print(f"Value Counts for {'col'} : ")
    display(df[col].value_counts())
```

## Univariate Analysis

```python
for i, predicator in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predicator)
```

## Bivariate Analysis

```python
#based on gender
for i, predicator in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predicator, hue='GenderId')
```

```python
#based on nationality
for i, predicator in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df, x=predicator, hue='Nationality')
```

```python
# histplot of value counts for different occupations

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

plt.figure(figsize=(30,30))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

## HeatMaps

```python
corr = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(corr, annot=True, cmap='crest', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA:

The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency
Account" indicating that customers who maintain high balances in one account type Often hold substantial amount/funds across Other
accounts as well.
